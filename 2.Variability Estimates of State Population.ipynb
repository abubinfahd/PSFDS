# Variability Estimates of State Population

## - Standard Deviation
## - Inter Quartile Range
## - Robust Median Absolute Deviation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load data

```python
df = pd.read_csv("D:/PSFDS/datasets/uspm.csv")
df.head()
```

```python
# Standard deviation
df['Population'].std()
```

```python
# Inter quartile range
df['Population'].quantile(0.75) - df['Population'].quantile(0.25)
```
