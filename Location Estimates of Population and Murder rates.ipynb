## Compute the mean, trimmed mean and median for the data

```python
# import necessary library
import pandas as pd
import numpy as np
from scipy.stats import trim_mean
```

```python
df = pd.read_csv("C:/Users/Abu Bin Fahd/Desktop/SFDS/datasets/uspm.csv")
df
```

```python
#mean
pop_mean = df['Population'].mean()
pop_mean
```

```python
#trimmed mean

pop_trim_mean = trim_mean(df['Population'], 0.1)
pop_trim_mean
```

```python
#werighted mean
wtd_mean = np.average(df['Murder rate'], weights=df['Population'])
wtd_mean
```

```python
#median
pop_med = df['Population'].median()
pop_med
```
